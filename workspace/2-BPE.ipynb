```python
with open("Data/combined_text.txt", "r") as f:
    text_sequence = f.read()

len(text_sequence)
```

```python
#Pour rajouter le minbpe dans le path
import sys
sys.path.append('..')
```

```python
from minbpe import BasicTokenizer

tokenizer = BasicTokenizer()
tokenizer.train(text_sequence, vocab_size=2048)
```

```python
vocab = tokenizer.vocab
vocab
```

```python
tokenizer.encode("Salut bebou")
```

```python
tokenizer.decode([83, 321, 117, 258, 2021])
```

```python
max_vocab_id = list(tokenizer.vocab.keys())[-1]
tokenizer.special_tokens = {
    "<|startoftext|>": max_vocab_id + 1,
    "<|separator|>": max_vocab_id + 2,
    "<|endoftext|>": max_vocab_id + 3,
    "<|unk|>": max_vocab_id + 4
}
```

```python
len(tokenizer.encode(text_sequence))
```

```python
tokenizer.save(file_prefix="Output/Tokenizer/my_tokenizer")
```
